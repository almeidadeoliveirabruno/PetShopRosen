```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
import time

# Configurações do Chrome
chrome_options = Options()
chrome_options.add_argument("--start-maximized")  # abre maximizado
# chrome_options.add_argument("--headless")  # rodar sem abrir janela

# Caminho para o chromedriver (ajuste conforme seu ambiente)
chromedriver_path = "C:/caminho/para/chromedriver.exe"
service = Service(chromedriver_path)

# Inicializa o driver
driver = webdriver.Chrome(service=service, options=chrome_options)

try:
    # Abre o Google
    driver.get("https://www.google.com")
    
    time.sleep(1)  # espera simples para carregar a página
    
    # Localiza a barra de pesquisa
    search_box = driver.find_element(By.NAME, "q")
    search_box.send_keys("Selenium Python tutorial")
    search_box.send_keys(Keys.RETURN)  # Pressiona Enter
    
    time.sleep(2)  # espera resultados carregarem
    
    # Pega o primeiro resultado
    first_result = driver.find_element(By.CSS_SELECTOR, "h3")
    print("Primeiro resultado:", first_result.text)
    
    # Opcional: clicar no primeiro resultado
    first_result.click()
    
    time.sleep(3)  # visualizar resultado
finally:
    # Fecha o navegador
    driver.quit()
```
